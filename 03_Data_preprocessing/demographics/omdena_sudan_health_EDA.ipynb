```python
file_path = "/content/drive/MyDrive/Omdena-Health_Sudan/demographicsAlt.csv"
```

```python
import pandas as pd
```

```python
data = pd.read_csv(file_path)
```

```python
data.info()
```

```python
# drop 'Unnamed: 0' column
data.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
# rename 'Region, subregion, country or area *' column
data.rename(columns={"Region, subregion, country or area *": "Region_country_area"}, inplace=True)
```

```python
# categorial columns
cat_cols = ["Region_country_area","Location code", "ISO2 Alpha-code","Type","Parent code", "indicator",
           "gender","quintile", "ages"]
```

```python
# how many different values for each categorial column ?
for col in cat_cols:
  print(f"{col} : {data[col].nunique()}")
```

```python
# genrate a sample from the unique values of 'indicator column
import random
random.sample(list(data["indicator"].unique()), 10)
```

```python
# get indicator related to health
health_indicators = []
indicator_list = list(data["indicator"].unique())
for indic in indicator_list:
  try:
    if "health" in indic:
      health_indicators.append(indic)
  except:
    pass
```
